### 단계 구분도(Choropleth)
- 서울시 구별 CCTV 설치대수

```python
import json, folium
import pandas as pd
```

- 데이터 준비

```python
df = pd.read_csv('data/cctv.csv', index_col='구별')
df.head()
```

```python
file = 'data/seoul_geo_simple.json'
geo_data = json.load(open(file, encoding='utf-8'))
```

- 단계 구분도

```python
map = folium.Map(location=[37.55,126.98], zoom_start=11, tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_data,
    data=df.CCTV댓수,        #단계구분도로 보여줄 데이터
    columns=[df.index,df.CCTV댓수],      # 데이터 프레임에서 추출할 column
    fill_color='PuRd',       # Color map
    key_on='feature.id'
).add_to(map)
map
```

- 자치구 이름 쓰기

```python
import numpy as np
def get_text_location(geo_str):
    gu_dict = {}
    for gu in geo_str['features']:
        for coord in gu['geometry']['coordinates']:
            geo = np.array(coord)
            gu_dict[gu['id']] = [np.mean(geo[:,1]), np.mean(geo[:,0])]
    return gu_dict
```

```python
gu_dict = get_text_location(geo_data)
gu_dict
```

```python
for gu_name in df.index:
    folium.map.Marker(
        location=gu_dict[gu_name],
        icon = folium.DivIcon(icon_size=(80,20), icon_anchor=(20,0),
            html=f'<div style="font-size: 10pt">{gu_name}</div>'
        )
    ).add_to(map)
map
```

```python
map = folium.Map(location=[37.55,126.98], zoom_start=11, tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_data,
    data=df.CCTV댓수,        #단계구분도로 보여줄 데이터
    columns=[df.index,df.CCTV댓수],      # 데이터 프레임에서 추출할 column
    fill_color='YlOrRd',       # Color map
    key_on='feature.id'
).add_to(map)
map
for gu_name in df.index:
    folium.map.Marker(
        location=gu_dict[gu_name],
        icon = folium.DivIcon(icon_size=(80,20), icon_anchor=(20,0),
            html=f'<div style="font-size: 10pt">{gu_name}</div>'
        )
    ).add_to(map)
title_html = '<h3 align="center" style = "font-size:20px">서울시 자치구별 CCTV 설치댓수</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```

- 최근증가율로 (Choropleth) 단계 구분도 만들기

```python
ru = folium.Map(location=[37.55,126.98], zoom_start=10, tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_data,
    data=df.최근증가율,
    columns=[df.index,'df.최근증가율'],
    fill_color='Blues',       
    key_on='feature.id',
).add_to(ru)
ru
for gu_name in df.index:
    folium.map.Marker(
        location=gu_dict[gu_name],
        icon = folium.DivIcon(icon_size=(80,20), icon_anchor=(20,0),
            html=f'<div style="font-size: 10pt">{gu_name}</div>'
        )
    ).add_to(ru)
title_html = '<h3 align="center" style = "font-size:20px">서울시 자치구별 CCTV 최근증가율</h3>'
ru.get_root().html.add_child(folium.Element(title_html))
ru
```
