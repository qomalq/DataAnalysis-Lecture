```python
df2 = pd.read_csv('data/경기도_주민등록인구집계현황.csv', encoding='euc-kr')
df2.head()
```

```python
df2 = df2[['행정구역구분명','행정구역명','총 인구수']]
```

```python
df2
```

```python
df2 =  df2[df2.행정구역구분명.str.contains('시군') | df2.행정구역구분명.str.contains('구')]
```

```python
gu_city = ['수원시', '고양시', '용인시', '성남시', '안산시', '안양시']
df2.행정구역명 = df2.행정구역명.str.strip()
drop_index = []
for index in df2[df2.행정구역구분명 == '시군'].index:
    if df2.행정구역명[index].split()[-1] in gu_city:
        drop_index.append(index)
drop_index
```

```python
df2.rename(columns={'행정구역구분명' : '시군구'}, inplace=True)
```
