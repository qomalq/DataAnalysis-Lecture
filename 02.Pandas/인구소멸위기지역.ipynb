```python
# 데이터 읽기
import numpy as np
import pandas as pd
dis_pop = pd.read_csv('data/주민등록인구집계현황.csv', encoding='euc-kr')
dis_pop.head()
```

```python
# 결측치 확인
dis_pop.isna().sum().sum()
```

```python
# 가임기 여성 인구수, 60세 이상 노령인구
dis_pop['가임기 여성 인구수'] = dis_pop['10~19세 (여)'] + dis_pop['20~29세 (여)']
dis_pop['60세 이상 노령인구'] = dis_pop['60~69세'] + dis_pop['70~79세'] + dis_pop['80~89세'] + dis_pop['90~99세'] + dis_pop['100세 이상']
```

```python
dis_pop['가임기 여성 인구수']
```

```python
dis_pop['60세 이상 노령인구']
```

```python
# 인구 소멸 비율
dis_pop['인구소멸 비율'] = dis_pop['가임기 여성 인구수'] / dis_pop['60세 이상 노령인구']
```

```python
dis_pop['인구소멸 비율']
```

```python
# 기초자치단체
dis_pop['기초자치단체'] = dis_pop['행정구역구분명']
```

```python
dis_pop = dis_pop['행정구역명,인구소멸 비율,기초자치단체'.split(',')]
dis_pop

```

```python
# 기초자치단체 읍면동 제거, 내림차순
index1 = dis_pop[dis_pop['기초자치단체'] == '읍면동'].index
dis_pop.drop(index1).sort_values(['인구소멸 비율']).head(10)
```
