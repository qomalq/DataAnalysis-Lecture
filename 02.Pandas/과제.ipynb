```python
import numpy as np
import pandas as pd
filename = 'data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv'
cctv = pd.read_csv(filename, encoding='euc-kr',thousands=',')
columns = '구별,CCTV댓수,2013년 이전설치된 CCTV,2013년,2014년,2015년,2016년,2017년,2018년,2019년,2020년,2021년,2022년'.split(',')
            #구분 -> 구별 / 총계 -> CCTV댓수
cctv = cctv[columns]
cctv.head()
```

```python
# 첫번째 행 계 삭제
cctv.drop([0], axis=0, inplace=True)
cctv.head()
```

```python
# 최근증가율 = 최근 3개년 cctv댓수의 합 / 이전 모든 년도의 CCTV 댓수
rec1 = cctv['2020년,2021년,2022년'.split(',')]
rec2 = cctv['2013년 이전설치된 CCTV,2013년,2014년,2015년,2016년,2017년,2018년,2019년'.split(',')]
rec1['3개년 합'] = rec1.sum(axis=1)
rec2['그 이전 모든 년도의 합'] = rec2.sum(axis=1)
# rec1['3개년 합'] = (cctv['2020년'] + cctv['2021년'] + cctv['2022년'])
```

```python
cctv['최근증가율'] = (rec1['3개년 합'] / rec2['그 이전 모든 년도의 합']).round(2) *100
```

```python
cctv = cctv[['구별','CCTV댓수','최근증가율',]]   
```

```python
cctv.구별 = cctv.구별.str.replace(' ','')
cctv.head()
```

```python
cctv.to_csv('data/cctv.csv', index=False)
df = pd.read_csv('data/cctv.csv')
df.head()
```

### 2. 서울시 인구 데이터를 다음과 같이 만들어 pop.csv 파일에 저장하세요.

  구별     인구수  내국인  외국인  고령자  외국인비율  고령자비율

3 용산구 237285 222953  14332   39070   6.04           16.47

```python
df = pd.read_csv('data/서울시인구.txt', sep='\t', encoding='utf-8',thousands=',')
df.head()
# df = pd.read_csv('data/서울시인구.txt', sep='\t',skiprows=2) -> 열 2줄 삭제하고 읽기
# df.head()
```

```python
# 불필요한 행열 삭제  기간	세대 인구.1	인구.2 인구.4	인구.5 인구.7	인구.8	세대당인구
df.drop([1], axis=0, inplace=True)
df.drop([2], axis=0, inplace=True)
df.drop(['기간','세대','인구.1', '인구.2', '인구.4', '인구.5', '인구.7', '인구.8', '세대당인구'], axis=1, inplace=True)
df.head()
```

```python
# 자치구 -> 구별 / 인구 ->인구수 / 인구.3 -> 내국인 / 인구.6-> 외국인/ 65세이상 고령자 -> 고령자 / 외국인비율 -> 외국인 //인구수 *100  / 고령자 비율 -> 고령자 // 인구수 *100
df.columns = '구별', '인구수', '내국인', '외국인', '고령자'
df.head()
```

```python
# 0	자치구	합계	한국인	등록외국인	65세이상고령자 삭제
df.drop([0], axis=0, inplace=True)
df.head()
```

```python
# str -> int로 변경
df['인구수'] = df['인구수'].str.replace(',','').astype(int)
df['내국인'] = df['내국인'].str.replace(',','').astype(int)
df['외국인'] = df['외국인'].str.replace(',','').astype(int)
df['고령자'] = df['고령자'].str.replace(',','').astype(int)
```

```python
# 외국인비율 -> 외국인 //인구수 *100  / 고령자 비율 -> 고령자 // 인구수 *100 추가\
df['외국인 비율'] =   ((df['외국인'] / df['인구수']) * 100).round(2)
df['고령자 비율'] =   ((df['고령자'] / df['인구수']) * 100).round(2)
df.index = df.index -2
```

```python
pop = df
pop
```

```python
pop.to_csv("pop.csv", index = False)
```
