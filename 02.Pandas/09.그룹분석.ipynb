#### 1. Pivot table

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]
df1 = pd.DataFrame(data, columns=columns)
df1
```

```python
df1.pivot('도시','연도','인구')
```

```python
df1.pivot_table(values='인구',index='도시', columns='연도')
```

```python
df1.pivot_table('인구','도시','연도')
```

```python
df1.pivot_table('인구',['지역','도시'],'연도')
```

###  2. 그룹 분석

```python
import seaborn as sns
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

##### 2.1 group by

- iris data

```python
iris.head()
```

```python
iris[50:53]
```

```python
iris.tail()
```

```python
# 품종별 항목의 개수
iris.groupby('species').count()
```

```python
# 품종별 항목의 평균
iris.groupby('species').mean()
```

```python
# 품종별 sepal_length(꽃받침 길이)의 평균 - 결과는 series
iris.groupby('species').mean()['sepal_length']
```

```python
iris[['sepal_length','species']].groupby('species').mean()
```

```python
iris.groupby('species')['sepal_length'].mean()
```

```python
# 품종별 항목의 평균과 표준편차
iris.groupby('species').agg(['mean','std'])
```

```python
# 개별 항목의 다양한 통계
iris.groupby('species')['sepal_width'].agg(['mean','std','min','max'])
```

- tips data

```python
print(tips.shape)
tips.head()
```

```python
tips['tip.pct'] = ((tips.tip / tips.total_bill)*100).round(2)
tips.tail()
```

```python
tips.describe()
```

```python
# 성별, 흡연여부, 팁의 비율의 평균,
tips.groupby(['sex','smoker'])[['tip.pct']].mean()
```

```python
# 성별, 흡연여부별 팁 비율의 데이터수, 평균, 최소, 최대
tips.groupby(['sex','smoker'])[['tip.pct']].agg(['count','mean','min','max'])
```

```python
tips.groupby(['sex','smoker','size'])[['tip.pct']].agg(['count','mean','min','max'])
```

```python
# 요일별, 시간대별, 팁 비율의 데이터 수, 평균, 최소, 최대
tips.groupby(['day','time'])[['tip.pct']].agg(['count','mean','min','max'])
```

##### 2.2 Pivot table

```python
# 성별, 흡연여부, 팁의 비율
tips.pivot_table('tip.pct', 'sex','smoker','mean')      # values, index, column, aggfunc
```

```python
# 합계 기능
# 성별, 흡연여부, 인원
tips.pivot_table('tip.pct','sex','smoker','count', margins=True, margins_name='계')
```

##### 2.3 group by vs pivot table

```python
# 성별 팁 비율의 평균 
tips.pivot_table('tip.pct','sex',aggfunc='mean')
```

```python
tips.groupby('sex')[['tip.pct']].mean()
```

```python
# 흡연 여부에 따른 평균 팁 비율과 최소 최대
tips.pivot_table('tip.pct','smoker',aggfunc=['mean','min','max'])
```

```python
tips.groupby('smoker')['tip.pct'].agg(['mean','min','max'])
```

```python
tips.groupby(['day'])['tip.pct'].agg(['mean','std'])
```

```python
tips.groupby(['time'])['tip.pct'].agg(['mean','std'])
```

```python
tips.groupby(['size'])['tip.pct'].agg(['mean','std'])
```
