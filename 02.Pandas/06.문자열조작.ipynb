- CCTV 데이터

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
filename = 'data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv'
cctv = pd.read_csv(filename, encoding='euc-kr')
cctv.head()
```

```python
# columns =['구분','총계',"2013년 이전설치된 CCTV",'2013년','2014년','2015년','2016년','2017년','2018년','2019년','2020년','2021년','2022년']
columns = '구분,총계,2013년 이전설치된 CCTV,2013년,2014년,2015년,2016년,2017년,2018년,2019년,2020년,2021년,2022년'.split(',')
```

```python
cctv = cctv[columns]
cctv.head()
```

```python
# 총계 데이터를 정수로 변환
cctv['총계'] = cctv.총계.str.replace(',','').astype(int)
cctv.head()
```

```python
# 나머지 colum도 변환
for col in columns[2:]:
    cctv['col'] = cctv[col].str.replace(',','').astype(int)
cctv.head()
```

```python
cctv.head()
```

```python
for i in cctv.index:
    print(cctv.구분[i], len(cctv.구분[i]))
```
