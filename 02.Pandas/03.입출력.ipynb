### 2. 데이터 입출력

```python
import numpy as np
import pandas as pd
```

- %%writefile

```python
%%writefile data/sample1.csv
c1,c2,c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

- CSv 파일 읽기

```python
df = pd.read_csv('data/sample1.csv')
df
```

```python
df.columns
```

- headerline 생략

```python
%%writefile data/sample2.csv
c1,c2,c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
df = pd.read_csv('data/sample2.csv', names = ['c1','c2','c3'])
df
```

```python
# index column 지정
df = pd.read_csv('data/sample1.csv', index_col='c1')
df
```

- tap or spaces로 구분된 파일

```python
%%writefile data/sample3.tsv
c1  c2  c3  c4
0.179181    -1.538472   1.347553   0.43381
1.024209    0.087307    -1.281997   0.49265
0.417899    -2.002308   0.255245   -1.10515
```

```python
df = pd.read_csv('data/sample3.tsv', sep= '\s+')    #여러개의 BLANK OR TAP: \S+
df
```

```python
df.to_csv('data/test.tsv', sep='\t', index=False)
```

- 여러 줄의 header
