```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
cctv = pd.read_csv('../02.Pandas/data/cctv_pop.csv')
cctv.head(3)
```

```python
sns.barplot(x=cctv.CCTV댓수,y=cctv.구별,data= cctv);
```

```python
ccv = cctv.구별.sort_index(ascending=False)
sns.barplot(x=cctv.CCTV댓수,y=ccv,data= cctv);
```

```python
cctv = cctv.sort_values(by=['CCTV댓수'], ascending=False) 
sns.barplot(x=cctv.CCTV댓수,y=cctv.구별,data=cctv);
```

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
dt = titanic.pivot_table('survived', 'sex','embark_town')
dt
```

```python
sns.heatmap(dt, cmap=sns.light_palette("Yellow", as_cmap=True), annot=True, fmt='f')
plt.title('sex,embark_town에 대한 생존률');
```

```python
sns.displot(x='age', kde=True, data=titanic);
```

```python
titanic.age.fillna(titanic.age.mean(),inplace=True)
sns.displot(x='age', kde=True, data=titanic);
```
