### 5.Pandas 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

- Line graght

```python
np.random.seed(2023)
df = pd.DataFrame(np.random.randn(100, 3),index=pd.date_range('2023-04-01', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df.head()
```

```python
df.plot(figsize=(5,3))
plt.title("Pandas의 plot method 사용 예")
plt.xlabel('일자')
plt.ylabel('데이터')
plt.show()
```

- Bar chart

```python
iris = sns.load_dataset('iris')
iris.sepal_length[:30].plot(kind='barh')
plt.title('Iris Seotsa 품종의 Sepal length');
```

```python
df2 = iris.groupby('species').mean()
df2
```

```python
df2.plot.bar(rot=0, figsize=(5,5))
plt.title('Iris 품종별 Feature의 평균');
```

```python
df2.columns.name = 'Feature'
df2.plot.bar(rot=0, figsize=(6,3))
plt.ylim(0,10)
plt.title('Iris 품종별 Feature의 평균');
```

- pie chart

```python
titanic = sns.load_dataset('titanic')
s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct='%.2f%%')
plt.title('선실별 승객수');
```

- histogram, kernerl density

```python
iris.plot.hist()
iris.plot.kde()         # scipy 모듈
plt.show()
```

- box plot

```python
iris[iris.species == 'setosa'].plot.box()
plt.title('iris setosa 품종의 각 feature 별 box plot')
```

```python
iris.boxplot(by ='species')
plt.tight_layout()
plt.suptitle('Iris 각 feature의 품종별 boxplot');
```
