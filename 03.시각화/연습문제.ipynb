```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
tip = sns.load_dataset('tips')
tip
```

```python
tip['팁의 비율'] = tip['tip'] / tip['total_bill']
```

```python
x , y = tip['day'] , tip['팁의 비율']
plt.title('요일별 팁의 비율')
plt.bar(x,y)
plt.xlabel('요일')
plt.ylabel('팁의 비율')
plt.show()
```

```python
x1, y1 = tip['size'] , tip['팁의 비율']
plt.title('인원수별 팁의 비율')
plt.bar(x1,y1)
plt.xlabel('인원수')
plt.ylabel('팁의 비율')
plt.show()
```

```python
ans = sns.load_dataset('anscombe')
ans1 = ans[0:11]
ans2 = ans[11:22]
ans3 = ans[22:33]
ans4 = ans[33:]
```

```python
ans1
```

```python
fig, axes = plt.subplots(2,2)
x = [ans1.x]
y = [ans1.y]
axes[0][0].plot(x,y,'o')
x = [ans2.x]
y = [ans2.y]
axes[0][1].plot(x,y,'o')
x = [ans3.x]
y = [ans3.y]
axes[1][0].plot(x,y,'o')
x = [ans4.x]
y = [ans4.y]
axes[1][1].plot(x,y,'o')
plt.show()
```

```python
fig, axes = plt.subplots(2,2)
x = [ans1.x]
y = [ans1.y]
axes[0][0].scatter(x,y)
x = [ans2.x]
y = [ans2.y]
axes[0][1].scatter(x,y)
x = [ans3.x]
y = [ans3.y]
axes[1][0].scatter(x,y)
x = [ans4.x]
y = [ans4.y]
axes[1][1].scatter(x,y)
plt.show()
```

```python
cctv = pd.read_csv('data/cctv.csv')
cctv = cctv[['구별','CCTV댓수']]
```

```python
cctv['구별'].head()
```

```python
plt.barh(cctv['구별'], cctv['CCTV댓수'])
plt.title('서울시 구별 CCTV 댓수')
plt.show()
```

```python
plt.barh(cctv['구별'].sort_index(), cctv['CCTV댓수'].sort_index())
plt.title('서울시 구별 CCTV 댓수')
plt.show()
```

```python
plt.barh(cctv['구별'].sort_values(), cctv['CCTV댓수'].sort_values())
plt.title('서울시 구별 CCTV 댓수')
plt.show()
```
