### 1. Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

#### Line Plot

```python
plt.plot([0,1,4,9])
plt.title('라인 플롯')
plt.grid()
plt.show()
```

### 스타일 지정

```python
plt.plot([0,1,4,9], 'rs--')     # red / r , square / s , 점선 --
plt.title("'rs--' 스타일의 플롯")
plt.grid()
plt.show()
```

```python
plt.plot([0,1,4,9], 'mo-.')    
plt.title("'mo-.' 스타일의 플롯")    
plt.plot([9,4,1,0], 'g^:')
plt.title("'g^:' 스타일의 플롯")
plt.grid()
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], "bo--",
         linewidth=5,markersize=15, markeredgecolor="g", markeredgewidth=5, markerfacecolor="r")
plt.title("스타일 적용 예")
plt.show()
```

### 그림 범위

```python
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.grid()
plt.show()
```

- 곡선 그림

```python
X = np.linspace(0,2*np.pi, 500)
Ysin = np.sin(X)
Ycos = np.cos(X)
plt.plot(X,Ysin), plt.plot(X,Ycos)
plt.title('Sine, Cosine 그래프')
plt.grid()
plt.show()
```

- Tick 설정

```python
plt.plot(X,Ysin), plt.plot(X,Ycos)
plt.title('Tick 설정')
plt.grid()
plt.xticks([0,np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
plt.yticks([-1,0,1])
plt.show()
```

```python
plt.plot(X,Ysin), plt.plot(X,Ycos)
plt.title('"LaTeX, 문자열로 tick label 정의"')
plt.xticks([0,np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
            [r'$0$',r'$\pi/2$', r'$\pi$', r'$3\pi$', r'$2\pi$'])
plt.grid()
plt.yticks([-1,0,1])
plt.show()
```

- 범례(legend)

```python
plt.plot(X,Ysin, label= 'Sine')
plt.plot(X,Ycos, label= 'Cosine')
plt.title('범례표시')
plt.xticks([0,np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
            [r'$0$',r'$\pi/2$', r'$\pi$', r'$3\pi$', r'$2\pi$'])
plt.yticks([-1,0,1])
plt.legend(loc= 0)   #upper center
plt.show()
```

- X, Y 축 Label

```python
plt.plot(X,Ysin, label= 'Sine')
plt.plot(X,Ycos, label= 'Cosine')
plt.title('X, Y 축 Label 표시')
plt.xticks([0,np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
            [r'$0$',r'$\pi/2$', r'$\pi$', r'$3\pi$', r'$2\pi$'])
plt.yticks([-1,0,1])
plt.legend(loc= 0)   #upper center
plt.xlabel('시간')
plt.ylabel('진폭')
plt.show()
```

- 그림의 크기 지정

```python
plt.figure(figsize=(10,6))
plt.plot(X,Ysin, label= 'Sine')
plt.plot(X,Ycos, label= 'Cosine')
plt.title('크기 지정(10,6)')
plt.xticks([0,np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
            [r'$0$',r'$\pi/2$', r'$\pi$', r'$3\pi$', r'$2\pi$'])
plt.yticks([-1,0,1])
plt.legend(loc= 0)   #upper center
plt.xlabel('시간')
plt.ylabel('진폭')
plt.grid()
plt.show()
```

- 그림 저장

```python
plt.figure(figsize=(10,6))
plt.plot(X,Ysin, label= 'Sine')
plt.plot(X,Ycos, label= 'Cosine')
plt.title('그림 저장', fontsize = 15)
plt.xticks([0,np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
            [r'$0$',r'$\pi/2$', r'$\pi$', r'$3\pi$', r'$2\pi$'])
plt.yticks([-1,0,1])
plt.legend(loc= 0)   #upper center
plt.xlabel('시간')
plt.ylabel('진폭')
plt.grid()
plt.savefig('data/삼각함수.png')
```
